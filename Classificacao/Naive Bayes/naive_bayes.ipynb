# Classificação com Naive Bayes (Multinomial)
Neste notebook, aplicamos o `MultinomialNB` para um problema clássico de Processamento de Linguagem Natural (NLP): **Análise de Sentimentos**.

O Multinomial Naive Bayes trabalha muito bem com contagens, então transformaremos nossos textos em matrizes numéricas (frequência de palavras) usando o `CountVectorizer`.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 1. Criando dados de exemplo (Textos e seus respectivos rótulos)
# 1 = Sentimento Positivo, 0 = Sentimento Negativo
textos_treino = [
    'Eu amei este produto, é maravilhoso e bonito',
    'Péssima qualidade, odiei, estragou rápido',
    'Muito bom, excelente custo benefício',
    'Não recomendo, quebrou no primeiro uso',
    'Perfeito, estou muito satisfeito com a compra',
    'Ruim demais, totalmente decepcionante'
]
y_treino = [1, 0, 1, 0, 1, 0]

# 2. Convertendo os textos em uma matriz de contagem de tokens (Bag of Words)
vectorizer = CountVectorizer()
X_treino = vectorizer.fit_transform(textos_treino)

# 3. Criando e treinando o modelo Multinomial Naive Bayes
modelo_nb = MultinomialNB()
modelo_nb.fit(X_treino, y_treino)

# 4. Testando o modelo com novos textos não vistos no treino
textos_teste = [
    'Produto excelente e muito bonito',
    'Horrível, não comprem, estragou',
    'Bom demais, amei'
]

# Lembre-se: Usamos apenas .transform() no teste para manter o vocabulário original
X_teste = vectorizer.transform(textos_teste)
previsoes = modelo_nb.predict(X_teste)

print("Classes previstas para os novos textos:")
for texto, previsao in zip(textos_teste, previsoes):
    sentimento = "Positivo" if previsao == 1 else "Negativo"
    print(f"-> '{texto}' | Sentimento: {sentimento}")
```
