# Classificação com Random Forest
Vamos aplicar o `RandomForestClassifier` em um dataset clássico de câncer de mama para ver como múltiplas árvores tomam decisões conjuntas.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# 1. Carregando os dados
dados = load_breast_cancer()
X = dados.data
y = dados.target # 0 = Maligno, 1 = Benigno

# 2. Separando em Treino e Teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Criando o modelo (100 árvores por padrão)
rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)

# 4. Treinando o modelo
rf_model.fit(X_treino, y_treino)

# 5. Avaliando o modelo
previsoes = rf_model.predict(X_teste)
print(f'Acurácia: {accuracy_score(y_teste, previsoes):.4f}\n')
print('Relatório de Classificação:\n', classification_report(y_teste, previsoes, target_names=dados.target_names))
```
