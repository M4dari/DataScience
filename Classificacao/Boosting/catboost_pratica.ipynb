# Classificação com CatBoost
Neste exemplo, vamos simular dados que contêm variáveis categóricas (texto) para mostrar a força do `CatBoostClassifier`, que não exige a transformação prévia dessas colunas.

```python
# !pip install catboost (descomente caso não tenha instalado)
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Criando um dataframe de exemplo (perfil de clientes e se compraram ou não)
dados = {
    'idade': [25, 45, 30, 50, 23, 40, 35, 60],
    'cidade': ['SP', 'RJ', 'BH', 'SP', 'BH', 'RJ', 'SP', 'BH'], # Var Categórica
    'profissao': ['Engenheiro', 'Médico', 'Professor', 'Médico', 'Estudante', 'Engenheiro', 'Professor', 'Médico'], # Var Categórica
    'comprou': [1, 1, 0, 1, 0, 1, 0, 1] # Target
}
df = pd.DataFrame(dados)
```

```python

X = df.drop('comprou', axis=1)
y = df['comprou']

# 2. Dividindo os dados
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Identificando as variáveis categóricas para o CatBoost
cat_features = ['cidade', 'profissao']
```

```python

# 4. Treinando o Modelo
modelo_cb = CatBoostClassifier(iterations=50, learning_rate=0.1, depth=3, verbose=0)
modelo_cb.fit(X_treino, y_treino, cat_features=cat_features)

# 5. Avaliando o modelo
previsoes = modelo_cb.predict(X_teste)
acuracia = accuracy_score(y_teste, previsoes)
print(f'Acurácia do CatBoost: {acuracia * 100:.2f}%')

# 6. Verificando a importância das features
importancias = modelo_cb.get_feature_importance()
for feature, imp em zip(X.columns, importancias):
    print(f'Feature: {feature} | Importância: {imp:.2f}')
```
