# Classificação com SVM (Support Vector Machine)
O módulo para classificação do SVM no Scikit-Learn chama-se `SVC` (Support Vector Classifier). Vamos avaliar o impacto de mudar o tipo de Kernel.

```python
from sklearn.svm import SVC
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 1. Criando um dataset não linear complexo (dados em formato de dois anéis concêntricos)
X, y = make_circles(n_samples=300, noise=0.1, factor=0.3, random_state=42)

# 2. Dividindo os dados
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Testando o SVM com Kernel Linear (Apenas retas)
svm_linear = SVC(kernel='linear')
svm_linear.fit(X_treino, y_treino)
acc_linear = accuracy_score(y_teste, svm_linear.predict(X_teste))
print(f'Acurácia SVM Kernel LINEAR: {acc_linear * 100:.2f}%')

# 4. Testando o SVM com Kernel RBF (Radial Basis Function - ideal para não lineares)
svm_rbf = SVC(kernel='rbf', C=1.0) # C é a penalidade (margem)
svm_rbf.fit(X_treino, y_treino)
acc_rbf = accuracy_score(y_teste, svm_rbf.predict(X_teste))
print(f'Acurácia SVM Kernel RBF: {acc_rbf * 100:.2f}%')

# Note como o kernel Linear fracassa terrivelmente (~50%) pois não consegue traçar uma linha reta 
# num círculo fechado, enquanto o RBF atinge 100% projetando para uma outra dimensão!
```
